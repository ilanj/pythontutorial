```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#create some data
x = np.linspace(0, 10, 100)
x
```

```python
#plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(x, x**6)
```

```python
fig, ax = plt.subplots()
x1 = [1,3,5,6,7]
y1 = [12, 13, 65, 77, 89]
ax.scatter(x1, y1)
```

```python
#sin wave
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```
